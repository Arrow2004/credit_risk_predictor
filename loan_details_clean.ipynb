```python
#faylni o'qish
import pandas as pd
df=pd.read_csv("./original_datasets/loan_details.csv")
df
```

```python
df.info()  #ustunlar haqida ma'lumot
```

```python
df.describe()  #raqamli ustunlarni tekshirish
```

```python
#loan_type ustunini tekshirish
print(df['loan_type'].describe()) 
print(df["loan_type"].unique())
```

```python
#loan_amount ustunini tekshirish
print(df['loan_amount'].describe()) 
print(df["loan_amount"].unique())
```

```python
# loan amount ustunidagi  $ va , belgilarini olib tashlash va float ga o'tkazish
df['loan_amount'] = df['loan_amount'].str.replace(r'[\$,]', '', regex=True).astype(float)

```

```python
# loan_type ustunini kichik harfga o'tkazish
df['loan_type'] = df['loan_type'].str.lower()

df
```

```python
# 'personal loan' ni 'personal' ga o'zgartirish
df['loan_type'] = df['loan_type'].replace('personal loan', 'personal')

# Tekshirish
print(df['loan_type'].value_counts())
```

```python
# loan_type ustunidagi kredit kartalarni 'creditcard' ga o'zgartirish
df['loan_type'] = df['loan_type'].replace({
    'credit card': 'creditcard',
    'cc': 'creditcard',
    'creditcard': 'creditcard'
})

# Tekshirish
print(df['loan_type'].value_counts())
```

```python
# Tozalangan faylni saqlash
df.to_csv("./cleaned_datasets/loan_details_cleaned.csv", index=False)

print("Fayl muvaffaqiyatli saqlandi!")
```
