```python
import pandas as pd

# CSV faylini yuklash
df = pd.read_csv("./original_datasets/financial_ratios.csv") 

# Tekshirish
print(type(df)) 
df.head(10)
```

```python
df.rename(columns={"cust_num": "customer_id"}, inplace=True)
```

```python
#"monthly_income" ustunini tozalash
df["monthly_income"] = (
    df["monthly_income"]
    .astype(str)                       # hammasini stringga aylantirish
    .str.replace("$", "", regex=False) # $ olib tashlash
    .str.replace(",", "", regex=False) # vergul olib tashlash
    .str.strip()                       # bo'shliqlarni olib tashlash
)

df["monthly_income"] = pd.to_numeric(df["monthly_income"], errors="coerce").fillna(0)
```

```python
#"existing_monthly_debt" ustunini tozalash
df["existing_monthly_debt"] = (
    df["existing_monthly_debt"]
    .astype(str)                       # hammasini stringga aylantirish
    .str.replace("$", "", regex=False) # $ olib tashlash
    .str.replace(",", "", regex=False) # vergul olib tashlash
    .str.strip()                       # bo'shliqlarni olib tashlash
)

df["existing_monthly_debt"] = pd.to_numeric(df["existing_monthly_debt"], errors="coerce").fillna(0)
```
