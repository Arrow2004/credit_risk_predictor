# Merging the files

```python
import pandas as pd

def gdrive_to_df(share_link, **kwargs):
  file_id = share_link.split('/')[5]
  download_url = f'https://drive.google.com/uc?id={file_id}'
  return pd.read_csv(download_url, **kwargs)

df1_link = "https://drive.google.com/file/d/1Cg1_khKyZNVERvfr2LSt63qwnY5yhtx3/view?usp=sharing"
df2_link = "https://drive.google.com/file/d/1WurfDDSzw5ZxlQZsIiSHjPX_xOi_Ayyh/view?usp=sharing"
df3_link = "https://drive.google.com/file/d/1Nf2m0k5v7ioH5ESbGu-aV22EAMhzHUhU/view?usp=sharing"
df4_link = "https://drive.google.com/file/d/10wwwLAveSAY1gh19r-wdFW-yg9ydX_9Z/view?usp=sharing"
df5_link = "https://drive.google.com/file/d/1j_p_3Btr6L4_LT-5mrxFYBahRa7CH0M3/view?usp=sharing"
df6_link = "https://drive.google.com/file/d/1Lq4aJrEJdSak47lozTSsIVk33RRdA6Lg/view?usp=sharing"

df1 = gdrive_to_df(df1_link)
df2 = gdrive_to_df(df2_link)
df3 = gdrive_to_df(df3_link)
df4 = gdrive_to_df(df4_link)
df5 = gdrive_to_df(df5_link)
df6 = gdrive_to_df(df6_link)
```

```python
def describe_with_missing(df, name):
  print(f"--- {name} ---")
  display(df.describe(include='all').T)
  print("\nMissing values per column:")
  print(df.isnull().sum())
  print("\n" + "="*50 + "\n")

describe_with_missing(df1, "application_metadata")
describe_with_missing(df2, "financial_ratios")
describe_with_missing(df3, "geographic_data")
describe_with_missing(df4, "loan_details")
describe_with_missing(df5, "demographics")
describe_with_missing(df6, "credit_history")
```

```python
def rename_first_column(df):
  df.columns = ['c_id'] + list(df.columns[1:])
  return df

df1 = rename_first_column(df1)
df2 = rename_first_column(df2)
df3 = rename_first_column(df3)
df4 = rename_first_column(df4)
df5 = rename_first_column(df5)
df6 = rename_first_column(df6)
```

```python
def print_columns(df, name):
  print(f"--- {name} columns ---")
  print(list(df.columns))
  print("\n" + "="*50 + "\n")

print_columns(df1, "application_metadata")
print_columns(df2, "financial_ratios")
print_columns(df3, "geographic_data")
print_columns(df4, "loan_details")
print_columns(df5, "demographics")
print_columns(df6, "credit_history")
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# Function to calculate Cramr's V
def cramers_v(x, y):
  confusion_matrix = pd.crosstab(x, y)
  chi2 = chi2_contingency(confusion_matrix)[0]
  n = confusion_matrix.sum().sum()
  return np.sqrt(chi2 / (n * (min(confusion_matrix.shape)-1)))

                # Dictionary to store correlations for discrete columns
discrete_corr = {}

for col in discrete_cols:
  discrete_corr[col] = cramers_v(merged_df[col].fillna('NA'), merged_df['default'])

discrete_corr_df = pd.DataFrame.from_dict(discrete_corr, orient='index', columns=['cramers_v'])
discrete_corr_df = discrete_corr_df.sort_values(by='cramers_v', ascending=False)

print("Discrete columns correlation with default (Cramr's V):")
print(discrete_corr_df)
```

```python
print("Pearson Correlation:")
pearson_corr = merged_df[continuous_cols].corr(method='pearson')
print(pearson_corr)

print("\nSpearman Correlation:")
spearman_corr = merged_df[continuous_cols].corr(method='spearman')
print(spearman_corr)

print("\nKendall Correlation:")
kendall_corr = merged_df[continuous_cols].corr(method='kendall')
print(kendall_corr)

plt.figure(figsize=(12,10))
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Pearson Correlation Heatmap")
plt.show()

plt.figure(figsize=(12,10))
sns.heatmap(spearman_corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Spearman Correlation Heatmap")
plt.show()

plt.figure(figsize=(12,10))
sns.heatmap(kendall_corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Kendall Correlation Heatmap")
plt.show()
```
