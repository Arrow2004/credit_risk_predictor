```python
#Kutubxonalarni import qilish
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Datasetni yuklash va umumiy ma'lumotlarni ko'rsatish
df = pd.read_csv("./original_datasets/demographics.csv")
df.describe()
```

```python
#Ustunlar haqida ma'lumot
df.info()
```

```python
#Ustun nomini almashtirish
df.rename(columns={"cust_id":"customer_id"},inplace=True)
```

```python
#Yo'qolgan malumotlarni qidirish
placeholders = ['NULL', 'NA', 'N/A', 'nan', r'\N']
print("\nCheck Placeholders:\n",(df.isin(placeholders)).sum())
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# "annual_income" ustunini tozalash va raqamli formatga o'tkazish
df["annual_income"] = (
    df["annual_income"]
    .astype(str)                       # hammasini stringga aylantirish
    .str.replace("$", "", regex=False) # $ olib tashlash
    .str.replace(",", "", regex=False) # vergul olib tashlash
    .str.strip()                       # bo'shliqlarni olib tashlash
)

df["annual_income"] = pd.to_numeric(df["annual_income"], errors="coerce").fillna(0)
df.info()
```

```python
#"eployment_length" ustunidagi yo'qolgan qiymatlarni o'rtacha bilan to'ldirish
df['employment_length'] = df['employment_length'].fillna(df['employment_length'].mean())
```

```python
#emloyment_type ustunidagi noyob qiymatlarni ko'rish
print(df['employment_type'].unique())
```

```python
#qiymarlarni standartlashtirish
mapping = {
    'Full-time': 'Full Time',
    'FULL_TIME': 'Full Time',
    'Full Time': 'Full Time',
    'Fulltime': 'Full Time',
    'Part Time': 'Part Time',
    'Self Employed': 'Self Employed',
    'FT': 'Full Time',
    'Self Emp': 'Self Employed',
    'Contractor': 'Contractor',
    'Self-employed': 'Self Employed',
    'SELF_EMPLOYED': 'Self Employed',
    'Contract': 'Contractor',
    'PART_TIME': 'Part Time',
    'CONTRACT': 'Contractor',
    'PT': 'Part Time',
    'Part-time': 'Part Time'

}

df['employment_type'] = df['employment_type'].map(mapping)

```

```python
# Tozalangan faylni saqlash
df.to_csv("./cleaned_datasets/demographics_cleaned.csv", index=False)
```
