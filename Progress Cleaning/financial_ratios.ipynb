```python
import pandas as pd

# CSV faylini qaytadan yuklash
df = pd.read_csv("financial_ratios.csv")  # Fayl nomini to'g'ri yozing

# Tekshirish
print(type(df))  # <class 'pandas.core.frame.DataFrame'> bo'lishi kerak
df.head(10)
```

```python
df.rename(columns={"cust_num": "customer_id"}, inplace=True)
```

```python
df.head(10)
```

```python
df["monthly_income"] = (
    df["monthly_income"]
    .astype(str)                       # hammasini stringga aylantirish
    .str.replace("$", "", regex=False) # $ olib tashlash
    .str.replace(",", "", regex=False) # vergul olib tashlash
    .str.strip()                       # bo'shliqlarni olib tashlash
)

df["monthly_income"] = pd.to_numeric(df["monthly_income"], errors="coerce").fillna(0)
```
